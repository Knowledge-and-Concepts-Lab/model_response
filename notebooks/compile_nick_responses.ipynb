```python
import pandas as pd
import os
```

```python
social_concepts_dir = '../experiments/social/responses/'
```

```python
df = pd.read_csv(os.path.join(social_concepts_dir, 'response_1.csv'), header=None)
#ignore the third column
df = df.iloc[:,0:2]
```

```python
for i in range(2, 101):
    temp_df = pd.read_csv(os.path.join(social_concepts_dir, 'response_' + str(i) + '.csv'), header=None)
    # add a new column to the dataframe and insert temp_df.iloc[:, 1] into it
    df['response_' + str(i)] = temp_df.iloc[:, 1]
```

```python
df.to_csv(os.path.join(social_concepts_dir, '../compiled_responses.csv'), index=False)
```
